```python
inputs1 = data.iloc[:, :9]
inputs2 = data.iloc[:, 11:13]
inputs3 = data.iloc[:, 14:16]

inputs = inputs1.join(inputs2)
inputs = inputs.join(inputs3)

inputs = np.array(inputs)

outputs = data.iloc[:, -1:].values
```

```python
inputs
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct = ColumnTransformer([('one', OneHotEncoder(), [1, 2, 3, 4, 6, 7, 8, 12])], remainder = 'passthrough')
inputs = ct.fit_transform(inputs)
```

```python
inputs.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(inputs,outputs,test_size = 0.2, random_state = 0)
```

```python
print('Inputs shape ', inputs.shape)
print('Outputs shape ', outputs.shape)
print('x_train shape ', x_train.shape)
print('x_test shape ', x_test.shape)
print('y_train shape ', y_train.shape)
print('y_test shape', y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```
