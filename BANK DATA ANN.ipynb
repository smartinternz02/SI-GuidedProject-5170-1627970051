```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
```

```python
ct = ColumnTransformer([('one', OneHotEncoder(), [1,2,3,4,6,7,8,10,15])], remainder = 'passthrough')
inputs = ct.fit_transform(inputs)

cto = ColumnTransformer([('one', OneHotEncoder(), [0])], remainder = 'passthrough')
outputs = cto.fit_transform(outputs)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(inputs,outputs,test_size = 0.2, random_state = 0)
```

```python
print('Inputs Shape ', inputs.shape)
print('Outputs Shape ', outputs.shape)
print('x_train Shape ', x_train.shape)
print('x_test Shape ', x_test.shape)
print('y_train Shape ', y_train.shape)
print('y_test Shape', y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
# ANN MODELLING
```

```python
from tensorflow.keras.models import Sequential  
from tensorflow.keras.layers import Dense
```

```python
categorical = Sequential()
```

```python
# Input layer
categorical.add(Dense(units = 51, kernel_initializer = 'random_uniform', activation = 'relu'))
```

```python
# Hidden layer 1  
categorical.add(Dense(units = 102, kernel_initializer = 'random_uniform', activation = 'relu'))
```

```python
# Hidden layer 2
categorical.add(Dense(units = 102, kernel_initializer = 'random_uniform', activation = 'relu'))
```

```python
# Output layer 
categorical.add(Dense(units = 2, kernel_initializer = 'random_uniform', activation = 'softmax'))
```

```python
# Compiling model  
categorical.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
# Training the model 
categorical.fit(x_train, y_train, epochs = 150)
```

```python
# Testing Model with Test Set 
ypred = categorical.predict(x_test)
```
